```python
###write a program to define the grade of a student basedon percentages:
##75 above ---Distinction
##50-75 above ---Second class
##35-50 ----pass
##below 35 --- fail
num=int(input())
if(num>75):
    print("Distinction")
elif(51<=num<=75):
    print("second class")
elif(35<=num<=50):
    print("pass")
else:
    print("fail")
```

```python
###write a program to print
##case-1:all three inputs are equal(10,10,10---10)
##case-2:all three are different(10,20,30-----60)
##case-3:two inputs equal(10,10,20-----20)
n1=int(input())
n2=int(input())
n3=int(input())
if(n1==n2 and n2==n3 and n3==n1):
    print(n1)
elif(n1!=n2 and n2!=n3 and n3!=n1):
    print(n1+n2+n3)
elif(n1==n2 and n1!=n3):
    print(n3)
elif(n1==n3 and n1!=n2):
    print(n2)
elif(n2==n3 and n2!=n1):
    print(n1)
```
