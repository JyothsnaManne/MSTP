### scatterplot(for sepallength and petallength)

```python
x=list(df['sepal_length'].values)
y=list(df['petal_length'].values)
plt.scatter(x,y,color='k')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.title('Scatter plot')
plt.show()
```

### bargraph

```python
plt.bar(x,y,color='pink')
plt.show()
```

### jointplot(for sepallength and sepalwidth)

```python
sb.jointplot("sepal_length", "sepal_width",data=df , kind="scatter", space=0, color="k")
```

### facetgrid(for sepallength and sepalwidth for all species)

```python
s=sb.FacetGrid(df,col='sepal_length',row='sepal_width')
s=s.map(plt.hist,'species',color='k')
```
